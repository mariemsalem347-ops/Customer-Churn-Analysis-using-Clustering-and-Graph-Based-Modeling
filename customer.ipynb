```python
# ================================
# Import libraries
# ================================
import pandas as pd
import numpy as np

# ================================
# Load dataset
# ================================
df = pd.read_csv("Telco-Customer-Churn.csv")

# ================================
# Initial inspection
# ================================
print(df.head())
print(df.info())

# ================================
# Data Cleaning (basic only)
# ================================

# Convert TotalCharges to numeric safely
# (fixes spaces or wrong string values)
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors='coerce')

# Drop missing values resulting from conversion
df = df.dropna()

# Remove duplicates if any
df = df.drop_duplicates()

# ================================
# Feature Engineering (YOUR PART ONLY)
# ================================

# 1. Average Charges per month
# Shows normalized spending behavior
df["AvgCharges"] = df["TotalCharges"] / (df["tenure"] + 1)

# 2. Customer Lifetime Value
# Total economic value of a customer
df["CustomerValue"] = df["MonthlyCharges"] * df["tenure"]

# 3. Engagement Score
# Combines usage duration and spending
df["EngagementScore"] = df["tenure"] * df["MonthlyCharges"]

# 4. Tenure Grouping
# Converts numeric tenure into meaningful categories
def tenure_group(x):
    if x < 12:
        return "New"
    elif x < 36:
        return "Mid-term"
    else:
        return "Long-term"

df["TenureGroup"] = df["tenure"].apply(tenure_group)

# 5. Spending Level
# Categorizes customers based on monthly charges
def spending_level(x):
    if x < 50:
        return "Low"
    elif x < 100:
        return "Medium"
    else:
        return "High"

df["SpendingLevel"] = df["MonthlyCharges"].apply(spending_level)

# 6. Charges Ratio
# Detects inconsistency between total and monthly charges
df["ChargesRatio"] = df["TotalCharges"] / (df["MonthlyCharges"] + 1)

# 7. Services Count
# Measures how many services each customer uses
services = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

df["ServicesCount"] = df[services].apply(lambda x: (x == "Yes").sum(), axis=1)

# ================================
# Final check
# ================================
print(df.head())
print(df.shape)

# Save processed dataset for next team member
df.to_csv("processed_telco.csv", index=False)
```
